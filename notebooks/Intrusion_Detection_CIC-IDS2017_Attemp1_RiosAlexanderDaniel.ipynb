# Attack descriptions
- **FTP-Patator**: Brute-force attack attempting to guess valid FTP credentials by systematically trying username and password combinations.
- **SSH-Patator**: Brute-force attack targeting SSH services to gain unauthorized remote access.
- **DoS**: attacks are attempts to make a service or server unavailable by overwhelming it with excessive requests or connections from a single source.
    - **DoS Slowloris**: Denial of Service attack that keeps many HTTP connections open by sending partial requests very slowly to exhaust server resources.
    - **DoS Slowhttptest**: Another slow-rate DoS attack using incomplete HTTP POST requests to overload web servers.
    - **DoS Hulk**: Floods web servers with a massive volume of HTTP requests to quickly exhaust resources and cause downtime.
    - **DoS GoldenEye**: High-rate HTTP flooding attack similar to Hulk, designed to crash web services.
- **Heartbleed**: Exploits a vulnerability in OpenSSL to read sensitive information from the server’s memory.
- **Web Attacks (Brute Force, XSS, SQL Injection)**
    - **Brute Force**: Automated attempts to guess login credentials on web applications.
    - **XSS (Cross-Site Scripting)**: Injecting malicious scripts into web pages.
    - **SQL Injection**: Injecting malicious SQL statements to manipulate or extract data from databases.
- **Infiltration**: A scenario in which an internal machine becomes compromised and is used to download malicious files or leak sensitive data.
- **DDoS**: Distributed Denial of Service attack performed by multiple systems sending traffic simultaneously to overwhelm the target.
- **Botnet Attack**: A coordinated attack from a network of infected machines under the attacker’s control.
- **PortScan**: Reconnaissance activity that systematically probes network ports to identify open services and potential vulnerabilities.

### The problem with null and negative values

In most cases, all features are expected to be greater than to zero, since they typically describe metrics related to flow size, packet counts, and time intervals within network communications.

#### Flow duration, flow bytes per second and total bytes

$ F_B = \frac{T_B}{t_F}\left[ \frac{Bytes}{Second} \right] $

Where:
- $ F_B $: Flow bytes per second
- $ T_B $: Total bytes
- $ t_F $: Flow duration

Futhermore:
- $ T_B = TL_{fwd} + TL_{bwd}$

Being:
- $ TL_{fwd} $: Total length of forward packets
- $ TL_{bwd} $: Total length of backward packets

Flow duration can be zero if one packet only was trasmited.

Zero total bytes trasmitted with flow duration > 0 is technically possible.
For example:
- The client initiates a TCP connection (SYN).
- The server replies with a SYN-ACK.
- The client never completes the handshake.
- Or the handshake completes, but no data is transmitted (it closes with RST).
- Or in UDP, the host attempts communication but there is no response.

The capture tool (CICFlowMeter, Zeek/Bro) could register that flow because it detects control packets (SYN, ACK) even though no application payload is exchanged.

In these cases:
- `flow_duration` is measured between the first and the last packet observed (for example, retransmissions or keep-alives).
- `total_bytes = 0` because there is no application-level payload.

So, it is not impossible.

#### Flow duration
A flow duration of zero isn’t impossible. But if there is more than one packet, it is impossible to have a flow duration of zero.

*Precision Issues*

> Capture tools such as CICFlowMeter or Zeek/Bro can have timestamp precision in microseconds or milliseconds.
If two packets arrive within the same measured time unit, the recorded difference is 0.
This does not mean that no time actually passed, but rather that the clock resolution does not distinguish the interval.

```python
flow_duration_zero = df.flow_duration <= 0
len(df[flow_duration_zero])
```

If the flow duration is zero, then the total number of packets transmitted must also be zero or one.

```python
df["total_packets"] = df["total_fwd_packets"] + df["total_backward_packets"]
packet_gt_one = df.total_packets > 1
len(df[flow_duration_zero & packet_gt_one])
```

#### Handling negative values

```python
cols_lt_zero = df.columns[:-2][(df.describe().min() < 0).values]
df[cols_lt_zero].describe().T
```

Observations:
- `flow_iat_min` and `fwd_iat_min`: These measure the time interval between packets. They can never be negative, because timestamps always increase over time.
- `fwd_header_length`, `bwd_header_length` and `fwd_header_length.1`: Header length measures how many bytes the TCP/IP headers occupy in a packet.
Values in the billions (negative) indicate data corruption or calculation errors.
A negative header length is impossible:
    - Ethernet, IP, TCP are all binary fields with positive lengths.
- `init_win_bytes_forward` and `init_win_bytes_backward`: This represents the initial receive window size in the connection. Valid values are 0 or positive (up to 65535 in classic TCP). Some capture tools use -1 to mark `unknown` or `not available`.
- `min_seg_size_forward`: The “Minimum segment size” measures the smallest observed TCP segment size. It can never be negative (segments cannot have negative size).

```python
mask = ((df.flow_iat_min < 0) |
        (df.fwd_iat_min < 0) |
        (df.fwd_header_length < 0) |
        (df.bwd_header_length < 0) |
        (df.init_win_bytes_forward < -1) |
        (df.init_win_bytes_backward < -1) |
        (df.min_seg_size_forward < 0))

df[mask].label.value_counts()
```

```python
classes = df[mask].label.value_counts()
classes * 100 / df.label.value_counts().loc[classes.index]
```

Removing these records does not involve a huge impact overall, but for the records with the `Heartbleed` label and the `Infiltrarion` label, it does. Therefore, for these records we'll replace the negative values with the minimum value per class label.

```python
df.drop(df[mask & (df.label != "Heartbleed") & (df.label != "Infiltration")].index, axis=0, inplace=True)

cols_lt_zero = df.columns[:-2][(df.describe().min() < 0).values]
df[cols_lt_zero].describe().T
```

```python
def fill_lt_zero(df, limit):
    cols_lt_zero = df.columns[:-2][(df.describe().min() < 0).values]
    for col in cols_lt_zero:
        mask = df[col] < limit
        targets = df[mask].label.unique()
        for target in targets:
            min_value = df[~mask & (df.label == target)][col].min()
            min_value = 1 if np.isnan(min_value) else min_value
            df.loc[(df.label == target) & mask, col] = min_value
    return df

df = fill_lt_zero(df, limit=1)
```

## Optimizing memory

```python
def optimizing_dtypes(df):
    bytesInOneMB = 1048576
    print(f"From {round(df.memory_usage().sum()/bytesInOneMB, 2)}MB", end="")
    col_float64 = df.select_dtypes(include=["float64"]).columns.values
    if len(col_float64) > 0:
        df[col_float64] = df.select_dtypes(include=["float64"]).astype("float32", copy=False)

    col_int64 = df.select_dtypes(include=["int64"]).columns.values
    if len(col_int64) > 0:
        for col in col_int64:
            if df[col].unique().tolist() == [1, 0] or df[col].unique().tolist() == [0, 1]:
                df[col] = df[col].astype("int8", copy=False)
            else:
                df[col] = df[col].astype("int32", copy=False)
    print(f" reduced to {round(df.memory_usage().sum()/bytesInOneMB,2)}MB")
    return df

df = optimizing_dtypes(df)
```

## Checking higher correlations

```python
th = 0.9
plt.figure(figsize=(18, 16))
sns.heatmap(corr[abs(corr) > th], cmap="coolwarm", annot=False, center=0)
plt.title("Correlation between numerical variables");
```

## Remove the worst features

```python
corr_pair = {}
colx_corr_target = []
coly_corr_target = []
index = corr[abs(corr) > th].index

for i, colx in enumerate(index):
    index_aux = index[i:]
    for coly in index_aux:
        corr_value = corr[abs(corr) > th].loc[colx, coly]
        if ~np.isnan(corr_value) and (colx != coly):
            corr_pair[f"{colx} - {coly}"] = corr_value
            colx_corr_target.append(np.absolute(corr_pearson.label.loc[colx]))
            coly_corr_target.append(np.absolute(corr_pearson.label.loc[coly]))

df_corr_max = pd.DataFrame(data={"Correlation": corr_pair.values(), "Corr_1st_feature": colx_corr_target, "Corr_2st_feature": coly_corr_target}, index=corr_pair.keys()).sort_values(by="Correlation", ascending=False)
mask = df_corr_max.Corr_1st_feature >= df_corr_max.Corr_2st_feature
df_corr_max["Worst_Feature"] = [df_corr_max.index[i].split(" - ")[value] for i, value in enumerate(mask)]
df_corr_max.head(10)
```

```python
features = list(set(df_corr_max.Worst_Feature.tolist()))
df.drop(features, axis=1, inplace=True)
```

## Filtering the best features

```python
df_temp = df.copy()

for col in df_temp.select_dtypes(include="object"):
    df_temp[col] = df_temp[col].astype("category").cat.codes

corr_pearson = df_temp.corr(method="pearson")
corr_spearman = df_temp.corr(method="spearman")

mask = corr_pearson.label.abs().sort_values(ascending=False)[1:] >= 0.1
list_features_1 = mask[mask].index.tolist()

mask = corr_spearman.label.abs().sort_values(ascending=False)[1:] >= 0.1
list_features_2 = mask[mask].index.tolist()

full_list_features = list(set(list_features_1 + list_features_2))
df_filtered = df[full_list_features + ["label"]]
```

## Model comparison

```python
from sklearn.metrics import balanced_accuracy_score, f1_score, precision_score, recall_score, roc_auc_score

def get_scores(y_true, y_pred, y_pred_proba):
    return {
        "Balanced Accuracy": balanced_accuracy_score(y_true, y_pred),
        "F1 Score": f1_score(y_true, y_pred, average="weighted"),
        "Precision": precision_score(y_true, y_pred, average="weighted"),
        "Recall": recall_score(y_true, y_pred, average="weighted"),
        "ROC AUC": roc_auc_score(y_true, y_pred_proba, average="weighted", multi_class="ovr")}

models = {"Logistic Regression": log_reg,
          "CatBoost Classifier": cat_clf,
          "XGBoost Classifier": xgb_clf}

scores = []
for model_name, model in models.items():
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)
    score = get_scores(y_test, y_pred, y_pred_proba)
    scores.append(score.values())

scores = pd.DataFrame(scores, columns=score.keys(), index=models.keys())
scores.style.highlight_max(color = 'green', axis = 0).highlight_min(color = 'red', axis = 0)
```

## Saving the scaler, the label encoder and the model with CloudPickle

An important difference between `cloudpickle` and `pickle` is that `cloudpickle` can serialize a function or class **by value**, whereas `pickle` can only serialize it **by reference**.

**Serialization by reference** treats functions and classes as attributes of modules, and pickles them through instructions that trigger the import of their module at load time. **Serialization by reference** is thus limited in that it assumes that the module containing the function or class is available/importable in the unpickling environment. This assumption breaks when pickling constructs defined in an interactive session, a case that is automatically detected by `cloudpickle`, that pickles such constructs **by value**.

```python
import cloudpickle

with open(input_folder+'intrusion-detection-model.bin', 'wb') as f_out:
    cloudpickle.dump((scaler, le, xgb_clf), f_out)
```
